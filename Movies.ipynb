```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('imdb_top_1000.csv')
```

```python
df.info()
```

```python
df = df.rename({'Series_Title':'Movie Title', 
                'Star1':'Leading Role', 
                'Star2':'Supporting Role 1',
                'Star3':'Supporting Role 2',
                'Star4':'Supporting Role 3'
                }, axis=1)
df = df.drop(['Poster_Link'], axis=1)
df.head(2)
```

```python
df.isnull().sum()
```

```python
print(df['Gross'].head(1))
df['Gross'] = df['Gross'].str.replace(',', '')
print(df['Gross'].head(1))

df['Gross'] = df['Gross'].astype('float64')
df['Gross'] = df['Gross'].replace(np.nan, 0)

df['Gross'] = df['Gross'].astype(int)
```

```python
df.describe()
```

```python
df['Runtime'] = df['Runtime'].str.extract('(\d+)').astype(int)
df['Runtime'].dtype
```

```python
df['No of Movies Directed'] = df.groupby('Director').transform('size')
df
```

```python
df.to_csv('Top 1000 Movies.csv')
```
